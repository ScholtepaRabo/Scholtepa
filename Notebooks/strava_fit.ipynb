```python
import pandas as pd
from fitparse import FitFile

def read_fit_file_to_dataframe(file_path):
    # Create a FitFile object
    fitfile = FitFile(file_path)
    
    # Initialize an empty list to store records
    records = []

    # Iterate over each record in the FIT file
    for record in fitfile.get_messages('record'):
        record_data = {}
        for data in record:
            record_data[data.name] = data.value
        records.append(record_data)
    
    # Convert the list of records to a DataFrame
    df = pd.DataFrame(records)
    
    return df

# Replace 'path_to_your_fit_file.fit' with the actual path to your FIT file
file_path = r'..\Data\Nu_de_beentjes_omhoog_tot_zondag_.fit'  # Replace with the path to your FIT file
df = read_fit_file_to_dataframe(file_path)

# Display the DataFrame
print(df)
```

## adjust long from FIT file to normal coordinates


```python
import pandas as pd

# Function to convert 32-bit signed integers to normal coordinates
def convert_to_coordinates(value):
    return value / (2**31) * 180

# Apply the conversion to the DataFrame
df['latitude'] = df['position_lat'].apply(convert_to_coordinates)
df['longitude'] = df['position_long'].apply(convert_to_coordinates)

print(df)
```

# drop rows with missing values in long or lat

```python
import pandas as pd

# Assuming df1 is already defined and contains 'latitude' and 'longitude' columns

# Check for missing values in 'latitude' and 'longitude' columns
missing_values = df[['latitude', 'longitude']].isnull().sum()
print(f"Missing values before dropping rows:\n{missing_values}")

# Drop rows with missing values in 'latitude' and 'longitude' columns
df_cleaned = df.dropna(subset=['latitude', 'longitude'])

# Check for missing values again to confirm
missing_values_after = df_cleaned[['latitude', 'longitude']].isnull().sum()
print(f"Missing values after dropping rows:\n{missing_values_after}")

# Display the cleaned DataFrame
print(df_cleaned)
df_cleaned
```

```python
import webbrowser
import folium
import pandas as pd

# Create a map centered around the average latitude and longitude of both dataframes
map_center = [df_cleaned['latitude'].mean(), df_cleaned['longitude'].mean()]
mymap = folium.Map(location=map_center, zoom_start=13)

# Add the GPX data to the map with different colors and add distance labels every 5 kilometers with labels
folium.PolyLine(df_cleaned[['latitude', 'longitude']].values.tolist(), color='blue', label='Half Marathon').add_to(mymap)


# Save the map to an HTML file
map_file = "combined_gpx_map.html"
mymap.save(map_file)

# Open the map in a browser
webbrowser.open(map_file)
```
